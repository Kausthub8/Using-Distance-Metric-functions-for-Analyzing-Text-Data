```
from textDB import load_data
corpus = load_data('hobbies')
```

```
corpus.categories
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
docs = vectorizer.fit_transform(corpus.data)
labels = corpus.target
```

```
docs.shape
```

```
from yellowbrick.text import TSNEVisualizer

tsne = TSNEVisualizer(size=(600,300))
tsne.fit(docs,labels)
tsne.poof()
```

```
time =TSNEVisualizer(size=(600,300), metrics="euclidean")
tsne.fit(docs,labels)
tsne.poof()
```

```
distance_functions = [
"bray_curtis", "canberra", "chebyshev", "cityblock", "correlation", "cosine", "dice", "euclidean", "hamming", "jaccard", "kulsinski", "mahalanobis", "matching", "minkowski", "rogerstanimoto", "russellrao", "seuclidean", "sokalmichener", "sokalsneath", "sqeuclidean", "yule"
]
```

```
tnse = TSNEVisualizer(metric="cityblock", size=(600,300))
tnse.fit(docs,labels)
tnse.poof()
```

```
tnse = TSNEVisualizer(metric="braycurtis", size=(600,300))
tnse.fit(docs,labels)
tnse.poof()
```

```
tnse = TSNEVisualizer(metric="canberra", size=(600,300))
tnse.fit(docs,labels)
tnse.poof()
```

```
tnse = TSNEVisualizer(metric="cosine", size=(600,300))
tnse.fit(docs,labels)
tnse.poof()
```

```
distance_functions = ["sokalmichener", "rogerstanimoto", "hamming"]
for metric in distance_functions:
  tsne = TSNEVisualizer(metric=metric)
  tsne.fit(docs,labels)
  tsne.poof()
```

```
from sklearn.cluster import KMeans
clusters = KMeans(n_clusters = 5)
clusters.fit(docs)

tsne = TSNEVisualizer(size=(1000,600))
tsne.fit(docs, ["c{}".format(c) for c in clusters.labels_])
tsne.poof()
```
